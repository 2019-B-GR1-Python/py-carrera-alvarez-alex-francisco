## <center> Análisis benchmarking

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_HDD = pd.read_csv('D:/tmp/HDD_Benchmarks.csv')
```

```python
df_HDD.head()
```

# <center> El Disco Duro 

El Disco Duro es un dispositivo magnético que almacena todos los programas y datos de la computadora. Su capacidad de almacenamiento se mide en gigabytes (GB) y es mayor que la de un disquete (disco flexible). Suelen estar integrados en la placa base donde se pueden conectar más de uno, aunque también hay discos duros externos que se conectan al PC mediante un conector USB.



## <center> Discos Duros que sobresalen en el mercado

```python
df_1 = df_HDD[['nombre_hdd','disco_rank']]
df_10_ranking = df_1.sort_values(by='disco_rank',ascending = False).tail(5)
df_10_ranking.plot.barh(x='nombre_hdd', y='disco_rank')
plt.show()
```

## <center> Discos Duros en base a su raiting

```python
df_2 = df_HDD[['nombre_hdd','disco_rating']]
df_10_raiting = df_2.sort_values(by='disco_rating',ascending = True).head(5)
df_10_raiting.plot.barh(x='nombre_hdd', y='disco_rating')
plt.show()
```

## <center> Discos Duros cuyo precio es mas alto

```python
df_3 = df_HDD[['nombre_hdd','price']]
df_10_caros = df_3.sort_values(by='price',ascending = True).tail(10)
df_10_caros.plot.barh(x='nombre_hdd', y='price')
plt.show()
```

## <center> Discos Duros marca INTEL 

```python
df_4 = df_HDD[df_HDD['nombre_hdd'].str.contains("Intel")]
df_Intel_top10 = df_4[['nombre_hdd','disco_rank']].sort_values(by='disco_rank',ascending = False).tail(5)
df_Intel_top10.plot( x='nombre_hdd', y='disco_rank' ,figsize=(14,4))
plt.show()
```

## <center> Discos Duros GYGABYTE mejor puntuados

```python
df_5 = df_HDD[df_HDD['nombre_hdd'].str.contains("Gigabyte")]
df_gyga_top5 = df_5[['nombre_hdd','disco_rating']].sort_values(by='disco_rating',ascending = False).head(5)
df_gyga_top5.plot( x='nombre_hdd', y='disco_rating' ,figsize=(14,4))
plt.show()
```

# <center> Tarjetas Graficas 
Una tarjeta gráfica es un dispositivo instalado en una computadora que está compuesto de una unidad de procesamiento gráfico diseñada para ayudar a procesar y generar imágenes, especialmente gráficas 3D. Las tarjetas gráficas ayudan a aliviar el esfuerzo de procesamiento del procesador principal, y pueden contener su propia memoria para aliviar el esfuerzo de la RAM del sistema.

```python
df_GPU = pd.read_csv('D:/tmp/GPU_Benchmarks.csv')
```

```python
df_GPU.head()
```

## <center> Tarjetas integradas
Las tarjetas gráficas integradas o a bordo están incorporadas en la placa madre de una computadora y suelen ser menos potentes que las tarjetas instaladas en una bahía de expansión.

## <center> Tarjetas dedicadas
Las tarjetas dedicadas son tarjetas gráficas que se instalan en bahías de expansión PCI Express, AGP, PCI u otras, y tienen su propia memoria de video separada de la RAM principal del sistema.

## <center> Tarjetas graficas con mejor ranking

```python
df_7 = df_GPU[['nombre_vc','videocard_rank']]
df_10_rankingGPU = df_7.sort_values(by='videocard_rank',ascending = True).head(10)
df_10_rankingGPU.plot.barh(x='nombre_vc', y='videocard_rank')
plt.show()
```

## <center> Tarjetas graficas con la calificación mas baja

```python
df_8 = df_GPU[['nombre_vc','passmark_rank']]
df_8_ratingGPU = df_8.sort_values(by='passmark_rank',ascending = False).tail(10)
df_8_ratingGPU.plot.barh(x='nombre_vc', y='passmark_rank')
plt.show()
```

## <center> Tarjetas graficas GForce mas vendidas

```python
df_9 = df_GPU[df_GPU['nombre_vc'].str.contains("GeForce")]
df_gf_top20 = df_9[['nombre_vc','videocard_rank']].sort_values(by='videocard_rank',ascending = False).head(10)
df_gf_top20.plot(kind='pie', y = 'videocard_rank', autopct='%2.1f%%', figsize=(9,9),
startangle=45, shadow=False, labels=df_gf_top20['nombre_vc'], fontsize=10)
plt.show()
```

# <center> Procesadores
El procesador es el que se refiere a los diferentes tipos de artículos de sistemas informativos que forma parte de un microprocesador que es parte de un CPU o micro que es el cerebro de la computadora y de todos los procesos informativos desde los más sencillos hasta los más complejos.

```python
df_CPU = pd.read_csv('D:/tmp/CPU_Benchmarks.csv')
df_CPU.head()
```

## <center> Procesadores AMD mejores rankeados

```python
df_11 = df_CPU[df_CPU['cpu_name'].str.contains("AMD")]
df_amd_10 = df_11[['cpu_name','cpu_rank']].sort_values(by='cpu_rank',ascending = True).head(10)
df_amd_10.plot.barh(x='cpu_name', y='cpu_rank' ,figsize=(10,4))
plt.show()
```

## <center> Procesadores mas vendidos en el mercado

```python
df_12 = df_CPU[['cpu_name','cpu_rank']]
df_100_ranking = df_12.sort_values(by='cpu_rank',ascending = False).head(100)
num_AMD = len(df_100_ranking[df_100_ranking['cpu_name'].str.contains("AMD")].index)
num_Intel = len(df_100_ranking[df_100_ranking['cpu_name'].str.contains("Intel")].index)
AMD = ['AMD', num_AMD]
Intel = ['Intel', num_Intel]
Otros = ['Otros', 100-num_AMD-num_Intel]

df_resultados = pd.DataFrame([AMD, Intel, Otros], columns=['Fabricante','Num'])
df_resultados.plot(kind='pie', y = 'Num', autopct='%2.1f%%', figsize=(5,5),
startangle=45, shadow=False, labels=df_resultados['Fabricante'], fontsize=10)
plt.show()
```

## <center> Procesadores de mayor costo INTEL

```python
df_13 = df_CPU[df_CPU['cpu_name'].str.contains("Intel")]
df_Intel_10 = df_13[['cpu_name','price']].sort_values(by='price',ascending = False).head(5)
df_Intel_10.plot.barh(x='cpu_name', y='price', figsize=(5,5))
plt.show()
```

# <center> Memoria RAM
La memoria RAM es la memoria principal de un dispositivo donde se almacena programas y datos informativos. Las siglas RAM significan “Random Access Memory” traducido al español es “Memoria de Acceso Aleatorio”.

La memoria RAM es conocida como memoria volátil lo cual quiere decir que los datos no se guardan de manera permanente, es por ello, que cuando deja de existir una fuente de energía en el dispositivo la información se pierde. Asimismo, la memoria RAM puede ser reescrita y leída constantemente.

```python
df_RAM = pd.read_csv('D:/tmp/RAM_Benchmarks.csv')
```

## <center> Memorias Ram mejor latencia Kingston

```python
df_14 = df_RAM[df_RAM['memory_name'].str.contains("Kingston")]
df_latencia_10 = df_14[['memory_name','latency']].sort_values(by='latency',ascending = True).head(5)
df_latencia_10.plot.barh(x='memory_name', y='latency', figsize=(5,5))
plt.show()
```

## <center> Memorias Ram con la mejor lectura sin cache

```python
df_151 = df_RAM[['memory_name','read_uncached']]
df_10_lectura = df_151.sort_values(by='read_uncached',ascending = True).head(10)
df_10_lectura.plot.barh(x='memory_name', y='read_uncached', figsize=(5,5))
plt.show()
```

## <center> Memorias Ram mejor velocidad escritura

```python
df_15 = df_RAM[['memory_name','write_speed']]
df_10_RAM = df_15.sort_values(by='write_speed',ascending = True).head(10)
df_10_RAM.plot.barh(x='memory_name', y='write_speed', figsize=(5,5))
plt.show()
```
