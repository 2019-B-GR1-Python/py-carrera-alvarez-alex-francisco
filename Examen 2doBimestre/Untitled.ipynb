## Crear un Dataframe con 10 registros y 6 columnas llenas de nmeros randomicos del 1 al 10 o valores NaN. Luego llenar los valores NaN con 0

```python
import pandas as pd
import numpy as np

dummyarray = np.empty((10,6))
dummyarray[:] = np.nan

df = pd.DataFrame(dummyarray)
print(df)
```

```python
df_bool = pd.isnull(df)
print(df_bool)
```

```python
df_sin_nan = df.dropna(how='any')
print(df_sin_nan)
```

## Crear un Dataframe con 10 registros y 6 columnas llenas de nmeros randomicos del 1 al 10 y sacar la media, la mediana, el promedio

```python
e = np.random.randint(1,11, size=(10, 6))
e_dataframe = pd.DataFrame(e)
print(e_dataframe)
```

```python
media = e_dataframe.mean()
print(media)
```

```python
mediana = e_dataframe.median()
print(mediana)
```

```python
promedio = e_dataframe.mode()
print(promedio)
```
